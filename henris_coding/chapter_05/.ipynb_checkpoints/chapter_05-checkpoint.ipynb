## Extracting the Principal Components step by step

In this subsection, we will tackel the first four steps of a PCA:

1.  Standardise the data
2.  Constructing the covariance matrix.
3.  Obtaining the eigenvalues and eigenvectors of the covariance matrix.
4.  Sortig the eigenvalues by decreasing order to rand the eigenvectors.

Firstly, we will start by loading the Wine dataset that we have been working with in Chapter 4, *Building Good Traiing Sets:  Data Preprocessing*.

````python
import pandas as pd

df_wine = pd.read_csv("/home/henri/stuff/machine_learning/sebastian_raschka/henris_coding/chapter_05/wine_data.txt",
                      header=None)
````

Next we will process the Wine data into separate training and test subsets - using 70 percent and 30 percent of the data respectively, and then standardise them to unit variance.

````python
from sklearn.model_selection import train_test_split
X, y = df_wine.iloc[:, 1:].values, df_wine.iloc[:, 0].values
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    stratify=y,
                                                    random_state=0)

# Standardise the features
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)
````

### Constructing the Covariance Matrix.

The symmetric $d\times d$-dimensional covariance matrix, where $d$ is the number of dimensions in the dataset, stores the pairwise covariances between the different features.

The eigenvectors of a covariance matrix represent the principal components (the directions of the maximum variance), whereas the corresponding eigenvalues will define their magnitude.

In the following, we will use the ```linalg.eig``` function from NumPy to obtain the eigenparis of the Wine Covariance matrix:

````python
import numpy as np
cov_mat = np.cov(X_train_std.T)
eigen_vals, eigen_vecs = np.linalg.eig(cov_mat)
print("Eigenvalues: \n{}".format(eigen_vals))
````

### Total and Explained Variance

Before we collect the $k$ most informative eigenvectors, let us plot the *variance explained ratios* of the eigenvalues.

The variance explained ration of an eigenvalue $\lambda_j$ is simply the fraction of an eigenvalue $\lambda_j$ and the total sum of the eigenvalues:
$$
\mathrm{Explained\_variance\_ratio} = \frac{\lambda_j}{\sum_{j=1}^{d}\lambda_j}
$$

Using Numpy ```cumsum``` function, we can calculate the cumulative sum of explained variances, which we will then plot via Matplotlib's ```step``` function.

````python
tot = sum(eigen_vals)
var_exp = [(i / tot) for i in sorted(eigen_vals, reverse=True)]
cum_var_exp = np.cumsum(var_exp)

import matplotlib.pyplot as plt
plt.bar(range(1,14), var_exp, alpha=0.5, align='center', label='inidividual explained variance')
plt.step(range(1,14), cum_var_exp, where='mid', label='cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Prinicipal Component Index')
plt.legend(loc='best')
plt.show()
````

The resulting plot indicates that the first principal component alone accounts for approximately 40 percent of the variance.  Also, we can see that the first two principal components combined explain almost 60 percent of the variance in the dataset.

PCA is an unsupervised method which means that information about the class labels is ignored.
Variance mesaures the spread of the the values along a feature axis.

### Feature Transformation

We will now process with the last 3 steps to transform the Wine dataset onto the new principla component axes.  The remaining steps we are going to tackle in this section are the following ones:

- Select $k$ eigenvectors, which correspond to the $k$ largest eigenvalues, where $k$ is the dimensionality of the new feature subspace $(k \le d)$.
- Construct a projection matrix from the "top" $k$ eigenvectors.
- Transform the $d$-dimensional input dataset $\mathbf{X}$ using the projection matrix $\mathbf{W}$ to obtain the new $k$-dimensional feature subspace.

Or in less technical terms, we will sort the eigenparis by descending order of the eigenvalues, construct a projection matrix from the selected eigenvectors, and use the projection matrix to transform the data onto the lower-dimensional subspace.

We start by sorting the eigenparis by decreasing order of the eigenvalues.

````python
# Make a list of (eigenvalue, eigenvector) tuples
eigen_pairs = [(np.abs(eigen_vals[i]), eigen_vecs[:, i]) for i in range(len(eigen_vals))]

# Sort the tuples from high to low
eigen_pairs.sort(key=lambda k: k[0], reverse=True)
````

Next, we collect the two eigenvectors that correspond to the two largest eigenvalues to capture about 60 percent of the variance in this dataset.  Note that we only chose 2 eigenvectors for the purpose of illustration, sice we are going to plot the data via a 2-dimensional scatter plot later in this subsection.

In practice, the number of prinipal components has to be determined by a trade-off between computational efficiency and the performance of the classifier:

````python
W = np.hstack((eigen_pairs[0][1][:, np.newaxis],
               eigen_pairs[1][1][:, np.newaxis]))
# newaxis is used to increase the dimensionality of an existing array by one more dimension.
print('Matrix W:\n{}'.format(W))
````

Now let us us transform the entire $124\times 13$-dimensional training dataset onto the two principal components by calculating the matrix dot product:
$$
\mathbf{X^{'}} = \mathbf{X}\cdot\mathbf{W}
$$

````python
X_train_pca = X_train_std.dot(W)
````

Lastly, let us visualize the transformed Wine traiing set, now stored as an $124\times 2$ dimensional matrix, in a 2-dimensional scatter plot:

````python
colors = ['r', 'b', 'g']
markers = ['s', 'x', 'o']
for l, c, m in zip(np.unique(y_train), colors, markers):
    plt.scatter(X_train_pca[y_train==l, 0],
                X_train_pca[y_train==l, 1],
                c=c, label=l, marker=m)
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.legend(loc='lower left')
plt.show()
````

As we can see in the resulting plot, the data is more spread along the $x$-axis -- the first principal component -- than the second principal component ($y$-axis), which is consistnt with the explained variance ratio plot that we created in the previous subsection.  However, we can intuitively see that a linear classifier will likely be able to separate the classes well.

Although we encoded the class label informtion for the purpose of illustration in the preceding scatter plot, we have to keep in mind that PCA is an unsupervised technque that does not use any class label information.

## Principal Component Analysis in scikit-learn

We will now discuss how to use the ```PCA``` class implemented in scikit-learn.
The ```PCA``` class is another one of scikit-learn's transformer classes, where we first fit the model using the traiing data before we transform both the training data and the test dataset using the same model parameters.

Now let us use the ```PCA``` class from scikit-learn on the Wine training dataset, classify the transformed samples via logistic regression, and visualise the decision regions via the ```plot_decision_region``` runction that we defined in Chapter 2.

````python
from matplotlib.colors import ListedColormap

def plot_decision_regions(X, y, classifier, resolution=0.02):
    # setup the marker generator and color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    
    # plot the decision surface
    x1_min, x1_max = X[:,0].min() - 1, X[:,0].max() + 1
    x2_min, x2_max = X[:,1].min() - 1, X[:,1].max() + 1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                           np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    
    plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
    plt.xlim([xx1.min(), xx1.max()])
    plt.ylim([xx2.min(), xx2.max()])
    
    # plot the class samples
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y==cl, 0],
                    y=X[y==cl, 1],
                    alpha=0.6,
                    c=cmap(idx),
                    edgecolor='black',
                    marker=markers[idx],
                    label=cl)

from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
lr = LogisticRegression()
X_train_pca = pca.fit_transform(X_train_std)
X_test_pca = pca.transform(X_test_std)
lr.fit(X_train_pca, y_train)

plot_decision_regions(X_train_pca, y_train, classifier=lr)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(loc='lower left')
plt.show()
````

By executing the preceding code, we sould now see the decision regions for the training data reduced to 2 principal component axes.

For the sake of completeness, let's plot the decision regions of the logistic regression on the tranformed test dataset to see if it can separate the classes well:

````python
plot_decision_regions(X_test_pca, y_test, classifier=lr)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(loc='lower left')
plt.show()
````

After we plottted the decision regions for the test set by executing the preceing code, we can see that the logistic regression performs quite well on this small 2d feature subspace and only misclassifier very few samples in the test dataset.

If we are interested in the explained variance ratios of the different principal components, we can simply initialise the ```PCA``` class with the ```n_componenets``` parameter set to ```None```, so all principal components are kept and the explained variance ratio can then be accessed via the ```explained_variance_ratio_``` attribute:

````python
pca = PCA(n_components=None)
X_train_pca = pca.fit_transform(X_train_std)
pca.explained_variance_ratio_
````

## Supervised data compression via linear discriminant analysis.

Linear Discriminant Analysis (LDA) can be used as a technique for feature extraction to increase the computational efficiency and reduce the degree of overfitting due to the curse of dimensionality in non-regularised models.

The goal in LDA is to find the feature subspace that optimises class separability.  In the following sections, we will discuss the similariteis between LDA and PCA in more detail and walk through the LDA approach step by step.

### The inner workings of linear discriminant analysis.

Let us briefly summarise the main steps that are required to perform LDA:

1. Standardise the $d$-dimensional dataset ($d$ is the number of features).
2. For each class, compute the $d$-dimensional mean vector.
3. Construct the betwee-class scatter matrix $\mathbf{S}_B$ and the within-calss scatter matrix $\mathbf{S}_W$.
4. Compute the eigenvectors and corresponding eigenvalues of the matrix $\mathbf{S}_W^{-1}\mathbf{S}_B$.
5. Sort the eigenvalues by decreasing order to rank the corresponding eigenvectors.
6. Chosse the $k$ eigenvectors that correspond to the $k$ largest eigenvalues to contruct a $d\times k$-dimensional transformation matrix $\mathbf{W}$; the eigenvetors are the columns of this matrix.
7. Project the samples onto the new feature subspace using the transformation matrix $\mathbf{W}$.

### Computing the Scatter Matrices

Since we already standradised the features of the Wine dataset in the PCAsection at the beginning of this chapter, we can skip the first step and proceed with the claculation of the mean vectors, which we will use to construct the within-class scatter matrix and the between-class scatter matrix, respectively.  Each mean vector $\mathbf{m}_i$ stores the mean features $\mu_m$ with respect to the samples of class $i$:

$$
\mathbf{m}_i = \frac{1}{n_i}\sum_{\mathbf{x}\in D_i}^{c}\mathbf{x}_m
$$

which results in three mean vectors (look at page 157).

````python
np.set_printoptions(precision=4)
mean_vecs = []
for label in range(1,4):
    mean_vecs.append(np.mean(X_train_std[y_train==label], axis=0))
    print("MV {0}:\n{1}\n".format(label, mean_vecs[label-1]))
````

Using the mean vectorsw, we can now compute the within-class scatter matrix $\mathbf{S}_W$:
$$
\mathbf{S}_W = \sum_{i=1}^{c}\mathbf{S}_i
$$
This is calculated by summing up the individual scatter matrices $\mathbf{S}_i$ of each individual class $i$:
$$
\mathbf{S}_i = \sum_{\mathbf{X}\in D_i}^{c}(\mathbf{x} - \mathbf{m}_i)(\mathbf{x} - \mathbf{m}_i)^T
$$

````python
d = 13 # the number of features
S_W = np.zeros((d, d))

for label, mv in zip(range(1,4), mean_vecs):
    class_scatter = np.zeros((d, d))
    for row in X_train_std[y_train==label]:
        row, mv = row.reshape(d, 1), mv.reshape(d, 1)
        class_scatter += (row - mv).dot((row - mv).T)
    S_W += class_scatter
print("Within Class Scatter Matrix: {}x{}".format(S_W.shape[0], S_W.shape[1]))
````

The assumption that we are making when we are computing the scatter matrices, is that the class labels in the training set are uniformly distributed.  However, if we print the number of class labels, we see that this assumption is violated.

````python
print("Class label Distribution:  {}".format(np.bincount(y_train)[1:]))
````

Thus we want to scale the individual scatter matrices $\mathbf{S}_i$ before we sum them up as scatter matrix $\mathbf{S}_W$.  When we divide the scatter matrices by the number of class-samples, $n_i$, we can see that computing the scatter matrix is in fact the same as computing the covariance matrix $\Sigma_i$.  The covariance matrix is a normalised version of the scatter matrix:

$$
\Sigma_i = \frac{1}{n_i}\mathbf{S}_W
$$

````python
d = 13 # number of features
S_W = np.zeros((d, d))
for label, mv in zip(range(1, 4), mean_vecs):
    class_scatter = np.cov(X_train_std[y_train==label].T)
    S_W += class_scatter
print("Scaled within-class scatter matrix: {}x{}".format(S_W.shape[0], S_W.shape[1]))
````

After we computed the scaled within-class scatter matrix (or covariance matrix), we can move on to the next step and compute the between-class scatter matrix $\mathbf{S}_B$:

$$
\mathbf{S}_B = \sum_{i=1}^{c}n_i(\mathbf{m}_i - \mathbf{m})(\mathbf{m}_i - \mathbf{m})^T
$$

Here, $\mathbf{m}$ is the overall mean that is computed, icluding samples from all classes:

````python
mean_overall = np.mean(X_train_std, axis=0)
d = 13
S_B = np.zeros((d, d))

for i, mean_vec in enumerate(mean_vecs):
    n = X_train[y_train==i+1, :].shape[0]
    mean_vec = mean_vec.reshape(d, 1)  # make a column vector
    mean_overall = mean_overall.reshape(d, 1)  # make another column vector
    S_B += n * (mean_vec - mean_overall).dot((mean_vec - mean_overall).T)
print("Between-class scatter matrix: {}x{}".format(S_B.shape[0], S_B.shape[1]))
````

### Selecting linear discriminants for the new feature subspace

Here we solve the generalised eigenvalue problem of the matrix $\mathbf{S}_W^{-1}\mathbf{S}_B$:

````python
eigen_vals, eigen_vecs = np.linalg.eig(np.linalg.inv(S_W).dot(S_B))
````

After we computer the eigenpairs, we can now sort the eigenvalues in decreasing order:

````python
eigen_pairs = [(np.abs(eigen_vals[i]),eigen_vecs[:, i]) for i in range(len(eigen_vals))]
eigen_pairs = sorted(eigen_pairs, key=lambda k: k[0], reverse=True)
print('Eigenvalues in descending order:\n')
for eigen_val in eigen_pairs:
    print(eigen_val[0])
````

In LDA, the number of linear discriminants is at most $c-1$, where $c$ is the number of class labels, since the in-between scatter matrix is the sum of $c$ matrices with rank 1 or less.

Now let us plot the linear discriminants by decreasing eigenvalues similar to the explained variance plot that we created in the PCA section.  For simplicity we will content of the class-disciriminatory information __discriminability__.

````python
tot = sum(eigen_vals.real)

discr = [(i / tot) for i in sorted(eigen_vals.real, reverse = True)]
cum_discr = np.cumsum(discr)

plt.bar(range(1,14), discr, alpha=0.5, align="center", label="individual discriminability")
plt.step(range(1, 14), cum_discr, where='mid', label="cumulative discriminability")
plt.ylabel("discriminability ratio")
plt.xlabel("Linear discriminants")
plt.ylim([-0.1, 1.1])
plt.legend(loc='best')
plt.show()
````

As we can see in the resulting figure, the first two linear discriminants alone capture 100 percent of the useful information in the Wine training dataset.

Let us now stack the 2 most disciminative eigenvector columns to create the transformation matrix $\mathbf{W}$:

````python
W = np.hstack((eigen_pairs[0][1][:, np.newaxis].real,
               eigen_pairs[1][1][:, np.newaxis].real))

print("Matrix W:\n {}".format(W))
````

## Projecting the samples onto the new feature space.

Using the transformtion matrix $\mathbf{W}$ that we created in the previous subsection, we can now transform the training dataset by multiplying the matrices:

$$
\mathbf{X}^{'} = \mathbf{X}\mathbf{W}
$$

````python
X_train_lda = X_train_std.dot(W)

colors = ['r', 'b', 'g']
markers = ['s', 'x', 'o']

for l, c, m in zip(np.unique(y_train), colors, markers):
    plt.scatter(X_train_lda[y_train==l, 0],
                X_train_lda[y_train==l, 1]*(-1),
                c=c, label=l, marker=m)
plt.xlabel('LD 1')
plt.ylabel('LD 2')
plt.legend(loc='lower right')
plt.show()
````

As we can see in the resulting plot, the three nwine classes are now perfectly linearly separable in the feature subspace.

## Using kernel principal component analysis for nonlinear mappings.

Using kernel PCA, we will learn how to transform data that is not linearly separable onto a new, lower-dimensional subspace that is suitable for linear classifiers.

Using the kernel trick, we can compute the similarity between 2 high-dimension feature vectors in the original feature space.  We use the kernel trick to avoid calculating the pairwise dot products of the samples $\mathbf{x}$ under $\phi$ explicitly by using the kernel function $\kappa$ so that we don't need to calculate the eigenvectors explicitly.

$$
\kappa\left( \mathbf{x}^{(i)}, \mathbf{x}^{(j)} \right) = \phi\left( \mathbf{x}^{(i)} \right)^T \phi\left( \mathbf{x}^{(j)} \right)
$$

In other words, what we obtain after the kernel PCA are the samples already projected onto the respective components, rather than constructing a transformation matrix as in the standard PCA approach.  Basically, the kernl funcion can be understood as a function that calculates a dot product between two vectors and __Gives a measure of similarity.__

The most commonly used kernels are the polynomial kernel, the hyperbolic tangent (sigmoid) kernel, and the RBF (or Gaussian) kernel.

To summarise what we have learned so far, we can define the following 3 steps to implement an RBF kernel PCA:
1.  We compute the kernel (similarity) matrix $\mathbf{K}$.
2.  We center the kernel matrix $\mathbf{K}$ using the following equation.
3.  We collect the top $\mathbf{k}$ eigenvectors of the centered kernel matrix based on their corresponding eigenvalues, which are ranked by decreasing magnitude.  In contrast to standard PCA, the eigenvectors are not the principal Component axes, but the samples already procject onto these axes.

### Implementing a kernel principal component analysis in Python:

````python
from scipy.spatial.distance import pdist, squareform
from scipy import exp
from scipy.linalg import eigh
import numpy as np

def rbf_kernel_pca(X, gamma, n_components):
    """
    RBF kernel PCA implementation.
    
    Parameters:
    -----------
    X: {NumPy ndarray}, shape = [n_samples, n_features]
    
    gamma: float
    Tuning parameter of the RBF kernel
    
    n_components: int
    Number of principal components to return.
    
    Returns:
    --------
    X_pc {NumPy ndarray}, shape = [n_samples, k_features]
    The projected dataset.
    """
    
    # Calculate the pairwise squared Euclidean distances in the MxN dimensional dataset.
    sq_dists = pdist(X, 'sqeuclidean')
    
    # Convert the pairwise distances into a square matrix.
    mat_sq_dists = squareform(sq_dists)
    
    # Compute the symmetric kernel matrix
    K = exp(-gamma * mat_sq_dists)
    
    # Center the kernel Matrix
    N = K.shape[0]
    one_n = np.ones((N, N)) / N
    K = K - one_n.dot(K) - K.dot(one_n) + one_n.dot(K).dot(one_n)
    
    # Obtaining the eigenpairs from the centered kernl matrix.
    # Note that scipy returns them in ascending order.
    eigvals, eigvecs = eigh(K)
    eigvals, eigvecs = eigvals[::-1], eigvecs[:, ::-1]
    
    # Collect the top k eigenvectors (projected samples)
    X_pc = np.column_stack((eigvecs[:, i] for i in range(n_components)))
    
    return X_pc
````

### Projecting new data-points

In kernel PCA we obtained the eigenvector, $\mathbf{a}$, of the centered kernel matrix (not the covariance matrix), which means that those are the samples that are already projected onto the prinicipal componenet axis $\mathbf{v}$.  Thus, ir we want to project a new sample $\mathbf{x}^{'}$ onto this principal component axis, we'd need to compute the following:

$$
\phi\left( \mathbf{x}^{'} \right)^{T}\mathbf{v}
$$

Fortunately, we can use the kernel trick so that we don't have to calculate the projection $\phi\left( \mathbf{x}^{'} \right)^{T}\mathbf{v}$ explicitly.  However, it is worth noting that kernel PCA, in contrast to standared PCA, is a memory-based method, which means that we have to re-use the original training set each time to project new samples.  We have to calculate the pairwise RBF kernel (similarity) between each $i$th sample in the training dataset and the new sample $\mathbf{x}^{'}$:

$$
\phi\left( \mathbf{x}^{'} \right)^{T}\mathbf{v} = \sum_{i}a^{(i)}\phi\left( \mathbf{x}^{'} \right)^{T}\phi\left( \mathbf{x}^{(i)} \right)
=\sum_{i}a^{(i)}\kappa\left( \mathbf{x}^{'},\mathbf{x}^{(i)} \right)
$$

Here the eigenvectors $\mathbf{a}$ and eigenvalues $\lambda$ of the kernel matrix $\mathbf{K}$ satsify the following condition in the equation:
$$
\mathbf{Ka} = \lambda\mathbf{a}
$$

After calculating the similarity between the new samples and the samples in the training set, we have to normalise the eigenvector $\mathbf{a}$ by its eigenvalue.  Thus, let us modify the ```rbf_kernel_pca``` function that we implemented earlier so that it __also returns the eigenvalues of the kernel matrix.__

````python
from scipy.spatial.distance import pdist, squareform
from scipy import exp
from scipy.linalg import eigh
import numpy as np

def rbf_kernel_pca(X, gamma, n_components):
    """
    RBF kernel PCA implementation.
    
    Parameters:
    -----------
    X: {NumPy ndarray}, shape = [n_samples, n_features]
    
    gamma: float
    Tuning the parameter of the RBF kernel.
    
    n_components : int
    The number of principal components to return.
    
    Retruns:
    --------
    X_pc: {NumPy ndarray}, shape = [n_samples, k_features]
    Projected dataset
    
    lambdas: list
    Eigenvalues
    """
    
    # Calculate the pairwise squared Euclidean distances in the MxN dimensional dataset.
    sq_dists = pdist(X, 'sqeuclidean')
    
    # Convert the pairwise distances into a square matrix
    mat_sq_dists = squareform(sq_dists)
    
    # Compute the symmetrix kernel matrix
    K = exp(-gamma * mat_sq_dists)
    
    # Compute the symmetric kernel matrix
    N = K.shape[0]
    one_n = np.ones((N,N)) / N
    K = K - one_n.dot(K) - K.dot(one_n) + one_n.dot(K).dot(one_n)
    
    # Obtaining the eigenpairs from the centered kernel matrix.
    # scipy.linalg.eigh returns them in ascending order
    eigvals, eigvecs = eigh(K)
    eigvals, eigvecs = eigvals[::-1], eigvecs[:, ::-1]
    
    # Collect the top k eigenvectors (projected samples)
    alphas = np.column_stack([eigvecs[:, i] for i in range(n_components)])
    
    # Collect the corresponding eigenvalues
    lambdas = [eigvals[i] for i in range(n_components)]
    
    return alphas, lambdas
````

Now let us create a new half-moon dataset and project it onto a one-dimensional subsapce using the updated RBF kernel PCA implementation:

````python
X, y = make_moons(n_samples=100, random_state=123)
alphas, lambdas = rbf_kernel_pca(X, gamma=15, n_components=1)
print("alphas.shape: {}\n".format(alphas.shape))
print("lambdas:\n{}".format(lambdas))
````

To make sure that we have implemented the code for projecting new samples, let us assume that the 26$^{th}$ point from the half-moon dataset is a new data point $\mathbf{x}^{'}$, and our task is to project it onto this new subspace:

````python
x_new = X[25]
x_new

x_proj = alphas[25]
x_proj

def project_x(x_new, X, gamma, alphas, lambdas):
    pair_dist = np.array([np.sum((x_new - row)**2) for row in X])
    k = np.exp(-gamma * pair_dist)
    
    return k.dot(alphas / lambdas)
````

By executing the following code, we are able to reproduce the original projection.  Using the ```project_x``` function, we will be able to project any new data sample as well.  The code is as follows:

````python
x_reproj = project_x(x_new, X, gamma=15, alphas=alphas, lambdas=lambdas)
x_reproj
````

Lastly, let us visualise the projection on the first principal component:

````python
plt.scatter(alphas[y==0, 0], np.zeros((50)), color='red', marker='^', alpha=0.5)
plt.scatter(alphas[y==1, 0], np.zeros((50)), color='blue', marker='o', alpha=0.5)
plt.scatter(x_proj, 0, color='black', label='original projection of point X[25]', marker='s', s=100)
plt.scatter(x_reproj, 0, color='green', label='remapped point X[25]', marker='x', s=500)
plt.legend(scatterpoints=1)
plt.show()
````

As we can now also see in the following scatterplot, we mapped the sample $\mathbf{x}^{'}$ onto the first principal component correctly.

### Kernel Principal Component Analysis in scikit-learn

For our convenience, scikit-learn implements a kernel PCA class in ```sklearn.decomposition``` submodule.  The usage is similar to the standard PCA class, and we can specify the kernel via the ```kernel``` parameter:

````python
from sklearn.decomposition import KernelPCA
X, y = make_moons(n_samples=100, random_state=123)
scikit_kpca = KernelPCA(n_components=2, kernel='rbf', gamma=15)
X_skernpca = scikit_kpca.fit_transform(X)
````

To check that we get the results that are consistent with our own kernel PCA implementation, let's plot the transformed half-moon shape data onto the first 2 principal components:

````python
plt.scatter(X_skernpca[y==0, 0], X_skernpca[y==0, 1], color='red', marker='^', alpha=0.5)
plt.scatter(X_skernpca[y==1, 0], X_skernpca[y==1, 1], color='blue', marker='o', alpha=0.5)
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.show()
````

As we can see, the results of scikit-learn's ```KernelPCA``` are consistent with our own implementation.

### Summary
Using PCA, we projected data onto a lower-dimensional subspace to maximise the variance along the orthogonal feature axes, while ignoring the class labels.

LDA, in contrast to PCA, is a technique for supervised dimensionality reduction, which means that it considers class information in the training datasets to attempt to maximise the class-separability in a linear feature space.

Using the kernel trick and a temporary projection into a higher-dimensional feature space, you were ultimately able to compress the datasets consisting of nonlinear features onto a lower-dimensional subspace where the classes became linearly separable.
